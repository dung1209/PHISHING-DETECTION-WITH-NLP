```
import pandas as pd
import numpy as np
from transformers import BertTokenizer, TFBertModel
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
import joblib

# 1. Đọc và xử lý dữ liệu
df = pd.read_csv('/content/drive/MyDrive/spam.csv', encoding='ISO-8859-1')
df = df[['v1', 'v2']].dropna()  # Lấy cột v1 (nhãn) và v2 (văn bản)

# Tách Spam và Ham để cân bằng dữ liệu
df_ham = df[df['v1'] == 'ham']
df_spam = df[df['v1'] == 'spam']
df_spam_upsampled = resample(df_spam, replace=True, n_samples=len(df_ham), random_state=42)
df_balanced = pd.concat([df_ham, df_spam_upsampled])

# Tách văn bản và nhãn
texts = df_balanced['v2'].values
labels = np.array([0 if label == 'ham' else 1 for label in df_balanced['v1']])

# 2. Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)

# 3. Tokenizer từ BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Hàm encode dữ liệu
def encode_texts(texts, tokenizer, max_len):
    encoded = tokenizer.batch_encode_plus(
        texts,
        max_length=max_len,
        padding='max_length',
        truncation=True,
        return_tensors='tf'
    )
    return encoded['input_ids'], encoded['attention_mask']

max_length = 128
X_train_ids, X_train_masks = encode_texts(X_train, tokenizer, max_length)
X_test_ids, X_test_masks = encode_texts(X_test, tokenizer, max_length)

# 4. Tạo đặc trưng TF-IDF ở cấp ký tự
tfidf_vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(3, 6), max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train).toarray()  # Huấn luyện và chuyển sang mảng numpy
X_test_tfidf = tfidf_vectorizer.transform(X_test).toarray()

# Lưu TfidfVectorizer
joblib.dump(tfidf_vectorizer, '/content/drive/MyDrive/tfidf_vectorizer.pkl')

# 5. Xây dựng mô hình sử dụng BERT + TF-IDF
class BertLayer(tf.keras.layers.Layer):
    def __init__(self, bert_model_name, **kwargs):
        super(BertLayer, self).__init__(**kwargs)
        self.bert_model_name = bert_model_name
        self.bert_model = TFBertModel.from_pretrained(self.bert_model_name)

    def call(self, inputs):
        input_ids, attention_masks = inputs
        bert_output = self.bert_model(input_ids, attention_mask=tf.cast(attention_masks, dtype=tf.int32))
        return bert_output.last_hidden_state[:, 0, :]

    def get_config(self):
        config = super(BertLayer, self).get_config()
        config.update({'bert_model_name': self.bert_model_name})
        return config

    @classmethod
    def from_config(cls, config):
        return cls(config['bert_model_name'])

# Input layers cho BERT
input_ids = tf.keras.layers.Input(shape=(max_length,), dtype=tf.int32, name='input_ids')
attention_masks = tf.keras.layers.Input(shape=(max_length,), dtype=tf.int32, name='attention_masks')

# Input layer cho TF-IDF
tfidf_input = tf.keras.layers.Input(shape=(X_train_tfidf.shape[1],), dtype=tf.float32, name='tfidf_input')

# BERT output
bert_output = BertLayer('bert-base-uncased')([input_ids, attention_masks])

# Kết hợp BERT và TF-IDF
combined_features = tf.keras.layers.Concatenate()([bert_output, tfidf_input])

# Fully connected layers
dense_out = tf.keras.layers.Dense(64, activation='relu')(combined_features)
dropout_out = tf.keras.layers.Dropout(0.3)(dense_out)
output = tf.keras.layers.Dense(1, activation='sigmoid')(dropout_out)

# Định nghĩa mô hình
model = tf.keras.Model(inputs=[input_ids, attention_masks, tfidf_input], outputs=output)

# 6. Biên dịch mô hình
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Hiển thị kiến trúc mô hình
model.summary()

# 7. Huấn luyện mô hình với callbacks
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6)
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(
    [X_train_ids, X_train_masks, X_train_tfidf],
    np.array(y_train),
    validation_data=([X_test_ids, X_test_masks, X_test_tfidf], np.array(y_test)),
    epochs=10,
    batch_size=32,
    callbacks=[lr_scheduler, early_stopping]
)

# 8. Đánh giá mô hình
accuracy = model.evaluate([X_test_ids, X_test_masks, X_test_tfidf], np.array(y_test))
print(f"Độ chính xác trên tập kiểm tra: {accuracy[1]:.2f}")

# Saving the model
model.save('/content/drive/MyDrive/spam_classifier_model_tf.keras')

# 9. Trực quan hóa quá trình huấn luyện
plt.figure(figsize=(12, 5))

# Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss During Training')
plt.legend()

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy During Training')
plt.legend()

plt.show()
```
